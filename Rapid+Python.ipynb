```python
data = pd.read_csv('HAM-clean.csv', sep=';', encoding='utf-8')
```

```python
#proses casefolding = Merubah kata menjadi huruf kecil semua
def casefolding(tweet):
    tweet = tweet.lower()
    tweet = tweet.strip(" ")
    tweet = re.sub(r'[?|$|.|!Â²_:")/(-+,]', '', tweet)
    return tweet
data['casefolding'] = data['content'].apply(lambda x: casefolding(x))
```

```python
#proses tokenizing = Memisahkan kata per kata dalam satu kalimat
def token(tweet):
    nstr = tweet.split(' ')
    dat = []
    a = -1
    for hu in nstr:
        a = a+1
    if hu == '':
        dat.append(a)
    p = 0
    b = 0
    for q in dat:
        b = q-p
        del nstr[b]
        p = p+1
    return nstr
data['tokenize'] = data['casefolding'].apply(lambda x: token(x))
data.head(20)
```

```python
#proses stemming = Mengubah menjadi kata dasar
def stemming(tweet):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    do = []
    for w in tweet:
        dt = stemmer.stem(w)
        do.append(dt)
    d_clean = []
    d_clean = " ".join(do)
    print(d_clean)
    return d_clean
data['stemming'] = data['stopword'].apply(lambda x: stemming(x))
data.head(20)
```

```python
# MENGHAPUS KOLOM YANG TIDAK DIBUTUHKAN

data.drop(data.columns[[0,2,3,4]], axis = 1, inplace = True)
```

```python
data_clean = data
data_clean.head(20)
```

```python
data_clean = data_clean.astype({'values' : 'category'})
data_clean = data_clean.astype({'stemming' : 'string'})
data_clean.dtypes
```

```python
#proses TF-IDF = Pembobotan tiap-tiap kata yang muncul, semakin banyak kata yang muncul berarti bobotnya semakin besar
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()
text_tf = tf.fit_transform(data_clean['stemming'].astype('U'))
text_tf
```

```python
#Splitting Data = Pembagian jumlah data untuk data training dan untuk data testing model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(text_tf, data_clean['values'], test_size=0.2, random_state=123)
```

```python
#Pemodelan dengan algoritma Naive Bayes
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

clf = MultinomialNB().fit(X_train, y_train)
predicted = clf.predict(X_test)

print("Naive Bayes Accuracy :" , accuracy_score(y_test,predicted))
print("Naive Bayes Precision :", precision_score(y_test,predicted, average="binary", pos_label="negative"))
print("Naive Bayes Recall :", recall_score(y_test,predicted, average="binary", pos_label="negative"))
print("Naive Bayes f1_score :", f1_score(y_test,predicted, average="binary", pos_label="negative"))

model = MultinomialNB()
clf_cv = cross_val_score(model, X_train, y_train, cv=10)
print("Naive Bayes Accuracy Cross Validation:", clf_cv.mean())

print(f'confusion matrix:\n {confusion_matrix(y_test, predicted)}')
print('=================================================\n')
print(classification_report(y_test, predicted))
```

```python
#Pemodelan dengan algoritma Support Vector Machine
from sklearn import svm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel
#Train the model using the training sets
clf.fit(X_train, y_train)
#Predict the response for test dataset
predicted = clf.predict(X_test)


print("SVM Accuracy :" , accuracy_score(y_test,predicted))
print("SVM Precision :", precision_score(y_test,predicted, average="binary", pos_label="negative"))
print("SVM Recall :", recall_score(y_test,predicted, average="binary", pos_label="negative"))
print("SVM f1_score :", f1_score(y_test,predicted, average="binary", pos_label="negative"))

model = clf
clf_cv = cross_val_score(model, X_train, y_train, cv=10)
print("SVM Accuracy Cross Validation:", clf_cv)

print(f'confusion matrix:\n {confusion_matrix(y_test, predicted)}')
print('=================================================\n')
print(classification_report(y_test, predicted))
```

```python
#Pemodelan dengan algoritma KNN
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score


classifier = KNeighborsClassifier(n_neighbors=9)  
classifier.fit(X_train, y_train) 
#Predict the response for test dataset
predicted = classifier.predict(X_test)

print("KNN Accuracy :" , accuracy_score(y_test,predicted))
print("KNN Precision :", precision_score(y_test,predicted, average="binary", pos_label="negative"))
print("KNN Recall :", recall_score(y_test,predicted, average="binary", pos_label="negative"))
print("KNN f1_score :", f1_score(y_test,predicted, average="binary", pos_label="negative"))

model = classifier
clf_cv = cross_val_score(model, X_train, y_train, cv=10)
print("SVM Accuracy Cross Validation:", clf_cv.mean())

print(f'confusion matrix:\n {confusion_matrix(y_test, predicted)}')
print('=================================================\n')
print(classification_report(y_test, predicted))
```
