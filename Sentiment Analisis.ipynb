# PROSES CLEANING

```python
# MENGHILANGKAN TANDA @ DAN USERNAMENYA

def remove_pattern(tweet, pattern):
    r = re.findall(pattern, tweet)
    for i in r:
        tweet = re.sub(i, '', tweet)
    return tweet
df['remove_username'] = np.vectorize(remove_pattern)(df['content'], "@[\w]*")
df.head(50)
```

```python
# REMOVE ANGKA, HASHTAG, RT, URL DAN SIMBOL2

def remove(tweet):
    tweet = re.sub('[0-9]+', '', tweet)
    tweet = re.sub(r'\$\w\n*', '', tweet)
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    tweet = re.sub(r'https\S+', '', tweet)
    tweet = re.sub(r'#\S+', '', tweet)
    tweet = re.sub(r'\n', '', tweet)
    tweet = re.sub(r'[!@#$%^.,&*)(-?/:;]', '', tweet)
    return tweet

df['remove_clean'] = df['remove_username'].apply(lambda x: remove(x))
df.head(100)
```

```python
#proses tokenizing = Memisahkan kata per kata dalam satu kalimat
def token(tweet):
    nstr = tweet.split(' ')
    dat = []
    a = -1
    for hu in nstr:
        a = a+1
    if hu == '':
        dat.append(a)
    p = 0
    b = 0
    for q in dat:
        b = q-p
        del nstr[b]
        p = p+1
    return nstr
data['tokenizing'] = data['casefolding'].apply(lambda x: token(x))
data.head(20)
```

# PROSES TF-IDF

```python
#proses TF-IDF = Pembobotan tiap-tiap kata yang muncul, semakin banyak kata yang muncul berarti bobotnya semakin besar
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()
text_tf = tf.fit_transform(data_tweet['tweet'].astype('U'))
text_tf
```

# SPLITING DATA

```python
#Splitting Data = Pembagian jumlah data untuk data training dan untuk data testing model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(text_tf, data_tweet['label'], test_size=0.2, random_state=123)
```

# MODELLING

```python
#Pemodelan dengan algoritma Naive Bayes
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

clf = MultinomialNB().fit(X_train, y_train)
predicted = clf.predict(X_test)

print("Naive Bayes Accuracy :" , accuracy_score(y_test,predicted))
print("Naive Bayes Precision :", precision_score(y_test,predicted, average="binary", pos_label="NEGATIF"))
print("Naive Bayes Recall :", recall_score(y_test,predicted, average="binary", pos_label="NEGATIF"))
print("Naive Bayes f1_score :", f1_score(y_test,predicted, average="binary", pos_label="NEGATIF"))

model = MultinomialNB()
clf_cv = cross_val_score(model, X_train, y_train, cv=10)
print("Naive Bayes Accuracy Cross Validation:", clf_cv.mean())

print(f'confusion matrix:\n {confusion_matrix(y_test, predicted)}')
print('=================================================\n')
print(classification_report(y_test, predicted))
```

```python
#Pemodelan dengan algoritma Support Vector Machine
from sklearn import svm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel
#Train the model using the training sets
clf.fit(X_train, y_train)
#Predict the response for test dataset
predicted = clf.predict(X_test)


print("SVM Accuracy :" , accuracy_score(y_test,predicted))
print('=================================================\n')
print("SVM Precision Positive :", precision_score(y_test,predicted, average="binary", pos_label="POSITIF"))
print("SVM Precision Negative :", precision_score(y_test,predicted, average="binary", pos_label="NEGATIF"))
print('=================================================\n')
print("SVM Recall Positive :", recall_score(y_test,predicted, average="binary", pos_label="POSITIF"))
print("SVM Recall Negative :", recall_score(y_test,predicted, average="binary", pos_label="NEGATIF"))
print('=================================================\n')
print("SVM f1_score :", f1_score(y_test,predicted, average="binary", pos_label="NEGATIF"))

model = clf
clf_cv = cross_val_score(model, X_train, y_train, cv=10)
print("SVM Accuracy Cross Validation:", clf_cv.mean())

print(f'confusion matrix:\n {confusion_matrix(y_test, predicted)}')
print('=================================================\n')
print(classification_report(y_test, predicted))
```
